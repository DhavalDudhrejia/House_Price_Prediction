# 4)Drop address, date, postcode, YearBuilt, lattitude, longtitude columns 

```python
df1=df.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis=1)
```

```python
df1.columns
```

# 5)Find the count of null value in each column  

```python
df1.isnull().sum()
```

# 6)Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0 

```python
df1['Propertycount'].fillna(0,inplace=True)
df1['Distance'].fillna(0,inplace=True)
df1['Bedroom2'].fillna(0,inplace=True)
df1['Bathroom'].fillna(0,inplace=True)
df1['Car'].fillna(0,inplace=True)
```

```python
df1.isnull().sum()
```

#  7)Fill Null value of land size and bidding area columns with Mean 

```python
df1['Landsize']=df1['Landsize'].fillna(df1['Landsize'].mean())
```

```python
 df1['BuildingArea']=df1['BuildingArea'].fillna(df1['BuildingArea'].mean())
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace=True)
```

```python
df1.shape
```

# 8)Find the unique value in method column 

```python
df1.Method.unique()
```

# 9)Create a dummy data for categorical data 

```python
df1=pd.get_dummies(df1,drop_first=True)
```

# Part-2: Working with Model  

#  1)Create the target data and feature data where target data is price

```python
df1.head()
```

```python
X=df1.drop('Price',axis=1)
Y=df1['Price']
```

#  2)Create a linear regression model for Target and feature data 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=36)
```

```python
X_train.shape
```

```python
X_test
```

```python
regression_model=LinearRegression()
```

```python
regression_model.fit(X_train,Y_train)
```

```python
predicted_price=regression_model.predict(X_test)
predicted_price
```

```python
Y_test
```

```python
regression_model.score(X_test,Y_test)
```

```python
regression_model.score(X_train,Y_train)
```

# 3)Check if the model is overfitting or underfitting or it is accurate 

The accuracy of the model for training data and testing data is almost same. So, there is no overfitting or underfitting issue.

# 4)If the model is overfitting then apply ridge and lasso regression algorithms 

There is no overfitting issue but we can always apply ridge and lasso regression and find out if there can be significant difference in score.

## a)Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
ridge_reg= Ridge(alpha=50, max_iter=100,)
ridge_reg.fit(X_train, Y_train)
```

```python
ridge_reg.score(X_test, Y_test)
```

```python
ridge_reg.score(X_train, Y_train)
```

## b) Lasso regression

```python
lasso_reg = Lasso(alpha=50, max_iter=100)
lasso_reg.fit(X_train, Y_train)
```

```python
lasso_reg.score(X_test, Y_test)
```

```python
lasso_reg.score(X_train, Y_train)
```
